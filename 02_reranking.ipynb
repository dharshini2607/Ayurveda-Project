```python
from sentence_transformers import CrossEncoder
import numpy as np
from typing import List, Tuple
import torch
```

```python
## 2. LOAD MODEL (20M params, fast)
model = CrossEncoder('cross-encoder/ms-marco-MiniLM-L-2-v2')
```

```python
## 3. TEST DATA (sample from Member 3)
claim = "Alice grew up in a small village near the sea"
chunks = [
    "Alice lived in a quiet coastal village during childhood.",
    "The city was bustling with people and noise everywhere.",
    "Alice's home overlooked the ocean waves.",
    "Village life was simple but fulfilling.",
    "She moved to the big city at age 18.",
    "Coastal towns have unique culture.",
    "Alice hated the crowded urban environment.",
    "She spent her early years by the seaside.",
    "Urban life never suited her rural roots.",
    "The village had a small population of 500.",
    "Ocean views defined her childhood memories.",
    "She left the village reluctantly for city work.",
    "Small town girl at heart, always.",
    "City apartments felt claustrophobic after village life.",
    "Her first 18 years: pure coastal village living."
]
```

```python
## 4. RERANK FUNCTION (YOUR MAIN DELIVERABLE)
def rerank_claim_evidence(claim: str, chunks: List[str], top_k: int = 5):
    """
    Rerank chunks using cross-encoder.
    Returns: List of (chunk, relevance_score) sorted by score.
    """
    if not chunks:
        return []

    pairs = [[claim, chunk] for chunk in chunks]
    scores = model.predict(pairs)  # cross-encoder scores each pair.[web:11][web:68]

    ranked = sorted(zip(chunks, scores), key=lambda x: x[1], reverse=True)
    return ranked[:top_k]
```

```python
## 5. TEST IT
result = rerank_claim_evidence(claim, chunks, top_k=5)
print("🎯 TOP 5 RERANKED CHUNKS:")
for i, (chunk, score) in enumerate(result, 1):
    print(f"{i}. Score: {score:.3f}")
    print(f"   '{chunk[:80]}...'")
    print()
```
